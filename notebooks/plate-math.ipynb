# Unique weight combinations
- using all possible plate weight pairs and the fewest number of plates
- (2.5, 5, 10, 15, 25, 35, 45, 55) pairs

```python
from itertools import combinations

possible_plates = [2.5, 5, 10, 15, 25, 35, 45, 55]
possible_plates_count = len(possible_plates)
unique_sums = {}

for size in range(1, possible_plates_count + 1):
    combinations_of_size = combinations(possible_plates, size)
    for combination in combinations_of_size:
        temp_sum = 45
        for element in combination:
            temp_sum += int(element * 2)

        # Add if sum does not exist or generated combination has a smaller combination sum
        if temp_sum not in unique_sums or sum(combination) < sum(unique_sums[temp_sum]):
            unique_sums[temp_sum] = combination

sorted_unique_sums = dict(sorted(unique_sums.items()))
for unique_sum in sorted_unique_sums:
    print(f"{int(unique_sum)}\t {sorted_unique_sums[unique_sum]}")
```
